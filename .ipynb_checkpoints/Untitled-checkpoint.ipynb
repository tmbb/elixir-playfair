```python
import matplotlib
import jinja2
```

```python
template = jinja2.Template("""
defmodule Playfair.Config do
  defstruct {{ key1 }}: {{ value1 }},{% for (key, value) in next_pairs %}
            {{ key }}: {{ value }}{% if not loop.last %},{% endif %}{% endfor %}
end
""")

def to_elixir(term):
    if term is True:
        return "true"
    elif term is False:
        return "false"
    elif term == "auto":
        return ":auto"
    elif isinstance(term, str):
        return repr(term).replace("'", "\"")
    else:
        return str(term).replace("'", "\"")
    
def keep_key(key):
    return not (
        key.startswith('keymap.') or
        key.startswith('mathtext.') or
        key.startswith('pdf.') or
        (key == 'text_usetex') or
        key.startswith('backend') or
        key.startswith('text_latex') or
        key.startswith('_') or
        key.startswith('agg.') or
        key.startswith('animation.')
    )

raw_pairs = list(dict(matplotlib.rcParams).items())
pairs = [
    (key.replace('.', '_'), to_elixir(value))
    for (key, value) in raw_pairs
    if keep_key(key)
]

pair1 = pairs[0]
(key1, value1) = pair1
next_pairs = pairs[1:]

print(template.render(key1=key1, value1=value1, next_pairs=next_pairs))
```
